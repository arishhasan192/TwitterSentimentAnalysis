```python
df2=df.drop(['tweet_coord','airline_sentiment_gold','negativereason_gold','tweet_created','tweet_id'],axis=1)
df2=df[['airline','text','airline_sentiment']]
data=np.array(df2)
data.shape
```

```python
import nltk
from nltk.tokenize import word_tokenize
```

```python
words=[]
for i in range(len(data)):
    temp=[]
    for j in range(len(data[i])-1):
        if j==1:
            temp=temp+(word_tokenize(data[i][j]))
        else:
            temp.append(data[i][j])
    words.append([temp,data[i][2]])
words
```

```python
words_test=[]
for i in range(len(data_test)):
    temp=[]
    for j in range(len(data_test[i])):
        if j==1:
            temp=temp+(word_tokenize(data_test[i][j]))
        else:
            temp.append(data_test[i][j])
    words_test.append(temp)
words_test
```

```python
from nltk.corpus import stopwords
import string
stops=stopwords.words('english')+list(string.punctuation)
stops
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
from nltk.corpus import wordnet
def get_simple_pos(tag): #creating simple tags to pass into the lemmatizer
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
from nltk import pos_tag
```

```python
def clean_review(words):
    output_words=[]
    for w in words:
        if w not in stops and type(w)!=float:
            pos=pos_tag([w])
            clean_word=lemmatizer.lemmatize(w,get_simple_pos(pos[0][1]))
            output_words.append(clean_word.lower())
    return output_words
```

```python
documents = [(clean_review(word), category) for word, category in words]
```

```python
documents_testing = [(clean_review(word)) for word in words_test]
```

```python
documents_testing[0]
```

```python
documents[0]
```

```python
all_words = []
for doc in documents:
    all_words += doc[0]
```

```python
freq = nltk.FreqDist(all_words)                 #will retrurn a freq distribution object
common = freq.most_common(3000)
features = [i[0] for i in common] #choosing the top 3000 frequency words
```

```python
def get_feature_dict(words): #will return true/false if the word in present in the document or not
    current_features = {}
    words_set = set(words)
    for w in features:
        current_features[w] = w in words_set
    return current_features
```

```python
training_data = [(get_feature_dict(doc), category) for doc, category in documents] 
```

```python
testing_data=[(get_feature_dict(doc),'') for doc in documents_testing]
```

```python
testing_data[0]
```

```python
training_data[0]
```

```python
from nltk import NaiveBayesClassifier 
classifier = NaiveBayesClassifier.train(training_data)
```

```python
nltk.classify.accuracy(classifier, training_data)
```
